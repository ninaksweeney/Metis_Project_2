# Increasing Electric Vehicle Adoption through Machine Learning

The goal of this project is to predict consumer ratings of cars in order to inform design decisions of Electric Vehicle (EV) manufacturers and increase EV adoption. To start exploring this problem, I scraped car data from [Kelly Blue Book's website](https://www.kbb.com/car-finder/). The structure of the site can be seen below. Scraping yielded over 4,000 records of cars, including features like expert rating, drivetrain,  brand, start price, trunk capacity, etc. Our target feature is consumer rating, a 1-5 star rating (rounded to the tenth), calculated based on consumer reviews. 

## EDA and Modeling

Initial EDA did not reveal any strong correlations with the target variable. In the correlation heatmap below, you can see that `peak_horsepower` , `expert_rating`, `start_price` and `entertainment_specs` have some of the strongest relationships to consumer rating. Pairplots yielded more insight, but still no major breakthroughs. 

I also have dummy variables that could have an impact on prediction accuracy. These include features like brand, drivetrain, transmission type, and existence of a crash rating. For each feature, I subtracted the mean consumer rating where the feature=0 from the mean consumer rating where the feature =1. The below plot shows that some dummies did show a lower or higher mean consumer rating than the rest of the data. Though there was no difference in means larger than 0.5 stars, the most notable features are brands, especially those of luxury and sports cars (seen on the far right). 

An initial Linear Regression model trained on `['expert_rating','entertainment_specs', 'peak_horsepower', 'security_specs', 'dummy_brand_McLaren', 'dummy_brand_Ferrari', 'dummy_brand_Bentley']` yielded an **R^2 of 0.124** and an **RMSE of 0.328**. Moving forward I plan to optimize the model as much as possible.
